# Correlation & Relationship

### Correlation Matrix: Correlation between different features in the dataset

```python
correlation_matrix = iris[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']].corr()
print(correlation_matrix)
```

```python
# Flatten the Matrix
```

```python
correlation_unstacked = correlation_matrix.unstack()
print(correlation_unstacked)
```

```python
# Remove duplicates and self correlation
```

```python
correlation_unstacked = correlation_unstacked[correlation_unstacked.index.get_level_values(0) != correlation_unstacked.index.get_level_values(1)]
print(correlation_unstacked)
```

### specific 'pair of features' that shows the 'strongest correlation'

```python
correlation_strongest = correlation_unstacked.sort_values(ascending=False)
print(correlation_strongest.head(1))
```

## Compare correlation among different species

```python
correlation_species = iris.groupby('Species')[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']].corr()
print(correlation_species)
```

```python
print(correlation_species.columns)
```

```python
print(iris.columns)
```

```python
correlation_sp_unstacked = correlation_species.unstack()
print(correlation_sp_unstacked)
```

```python
print(correlation_sp_unstacked.columns.tolist())
```

```python
# Remove Self Correlations
```

```python
# correlation_sp_unstacked_filtered = correlation_sp_unstacked[correlation_sp_unstacked.columns.droplevel(0) != correlation_sp_unstacked.index.get_level_values(1)]
# print(correlation_sp_unstacked_filtered)
```

```python
# Check if correlation_sp_unstacked is a DataFrame or a groupby object
```

```python
print(type(correlation_sp_unstacked))
```

```python
# To reset 'Species' from row label to Columns after groupby removed 'Species' from Column
```

```python
correlation_sp_unstacked = correlation_species.unstack().reset_index()
print(correlation_sp_unstacked)
```

```python
print(correlation_sp_unstacked.columns)
```

```python
# correlation_sp_unstacked.columns = correlation_sp_unstacked['Species', 'Feature1', 'Feature2', 'Correlation']
```

```python
print(correlation_sp_unstacked.shape)
```

```python
print(correlation_sp_unstacked.columns)
```

```python
# Flatten MultiIndex Columns. Tuples are converted to Strings. 'Species' is retained 
```

```python
correlation_sp_unstacked.columns = ['_'.join(col).strip() if col[1] else col[0] for col in correlation_sp_unstacked.columns]
print(correlation_sp_unstacked.columns)
```

```python
# Rename the first column to Species
```

```python
# correlation_sp_unstacked.rename(columns={'Species_': 'Species'}, inplace=True)
```

```python
if 'Species_' in correlation_sp_unstacked.columns:
    correlation_sp_unstacked.rename(columns={'Species_': 'Species'}, inplace=True)
elif 'Species' not in correlation_sp_unstacked.columns:
    print("Warning: 'Species' column not found in the DataFrame!")
```

```python
# Remove Self Correlation
```

```python
valid_columns = [col for col in correlation_sp_unstacked.columns if '_' in col and col.split('_')[0] != col.split('_')[1]]
print(valid_columns)
```

```python
valid_columns.append('Species')
```

### Correlation of Feature Pairs among different Species

```python
# selected filtered columns only
```

```python
correlation_sp_unstacked_filtered = correlation_sp_unstacked[valid_columns]
print(correlation_sp_unstacked_filtered)
```
