```python
# Write a Pandas program to add, subtract, multiple and divide two Pandas Series. Sample Series: [2, 4, 6, 8, 10], [1, 3, 5, 7, 9]

s1 = pd.Series([2,4,6,8,10])
print(s1)
```

```python
s2 = pd.Series([1,3,5,7,9])
print(s2)
```

```python
s_addition = s1 + s2
print(s_addition)
```

```python
s_substraction = s1 -s2
print(s_substraction)
```

```python
s_multiplication = s1 * s2
print(s_multiplication)
```

```python
s_division = s1 / s2
print(s_division)
```
